```python
import pandas as pd
df1 = pd.DataFrame({"City":["New York", "Chicago", "Orland", "Baltimore"], "Temp":[21, 14,35,32]})
```

```python
df1
```

```python
df2 = pd.DataFrame({"City":["Chicago", "New York", "Orland", "DC"], "Humidity":[65,68,75, 85]})
```

```python
df2
```

```python
Weather = pd.merge(df1, df2, on= "City")
```

```python
Weather
```

```python
Weather1 = df1.merge(df2, on= "City")
```

```python
Weather1
```

```python
df4 = pd.merge(df1, df2, on= "City", how = "left")
```

```python
df4
```

```python
Weather = pd.merge(df1, df2, on= "City", how = "right")
```

```python
Weather
```

```python
Weather = pd.merge(df1, df2, on= "City", how = "outer")
```

```python
Weather
```

```python
Weather = pd.merge(df1, df2, on= "City", how = "outer", indicator = True)
```

```python
Weather
```

```python
Weather = pd.merge(df1, df2, on= "City", how = "left", indicator = True)
```

```python
Weather
```

```python
df3 = pd.DataFrame({"City":["New York", "Chicago", "Orland", "Baltimore"], "Temp":[23, 17, 28, 23], "Humidity":[87,78,82, 88]})
df3
```

```python
df5 = pd.merge(df4, df3, on= "City", suffixes= ("_Jan", "_Feb"))
df5
```

```python
df6 = pd.DataFrame({"City":["Chicago", "New York", "Orland", "DC"], "Temp":[25, 27, 22, 28], "Humidity":[65,68,75, 85]})
df6.set_index('City', inplace = True)
df6
```

```python
df7 = pd.DataFrame({"City":["New York", "Chicago", "Orland", "Baltimore"], "Temp":[23, 17, 28, 23], "Humidity":[87,78,82, 88]})
df7.set_index('City', inplace = True)
df7
```

```python
Weather = df6.join (df7, lsuffix = '_Jan', rsuffix = '_Feb', how = "outer")
Weather
```
